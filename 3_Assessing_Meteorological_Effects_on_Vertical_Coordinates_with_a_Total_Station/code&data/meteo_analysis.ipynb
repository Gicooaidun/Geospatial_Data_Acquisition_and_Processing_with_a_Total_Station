```python
import pandas as pd
from meteo import plot, plot_meteo
```

```python
PATH = ''
SENSOR = 'dft1' # dft1 or mws95
data_original = pd.read_csv(PATH+f'Data/Original_data/reinhard_ref.csv', index_col='Unnamed: 0')
data_original.index = pd.to_datetime(data_original.index)
data_corrected = pd.read_csv(PATH+f'Data/Corrected_data/reinhard_ref_vertical_offsets.csv', index_col='Unnamed: 0')
data_corrected.index = pd.to_datetime(data_corrected.index)
```

```python
# Plot original Reinhardt data
plot(data_original, ['Temperature', 'Pressure', 'Humidity'])
```

```python
# Plot corrected Reinhardt data
plot(data_corrected, ['Temperature', 'Pressure', 'Humidity'])
```

```python
# Example data from the meteo huts:
data_original_meteo = pd.read_csv(PATH+f'Data/Original_data/meteo_huts.csv')
data_original_meteo.index = pd.to_datetime(data_original_meteo['Datetime'])
data_corrected_meteo = pd.read_csv(PATH+f'Data/Corrected_data/meteo_huts_vertical_offsets.csv')
data_corrected_meteo.index = pd.to_datetime(data_corrected_meteo['Datetime'])
```

```python
# Select the data of the required meteo hut (e.g. Lab17):
# All available Meteo Huts: ['Lab17', 'Lab18', 'Lab19', 'Lab22', 'Lab24', 'Lab25', 'Lab26', 'Lab27']
STATION = 'Lab24'
SENSOR = 'BME280 H_Lab24'
meteo_hut_data = data_original_meteo.filter(regex=f'_{STATION}$')
meteo_hut_data.columns = [col.replace('_'+STATION, '') for col in meteo_hut_data.columns]
```

```python
# Plot original data
plot_meteo([meteo_hut_data], [SENSOR], data_original)
```

```python
STATION = 'Lab24'
SENSOR = 'BME280 H_Lab24'
meteo_hut_data = data_corrected_meteo.filter(regex=f'_{STATION}$')
meteo_hut_data.columns = [col.replace('_'+STATION, '') for col in meteo_hut_data.columns]
```

```python
# Plot corrected data
plot_meteo([meteo_hut_data], [SENSOR], data_original)
```
